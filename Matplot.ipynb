```python
import matplotlib.pyplot as matp
%matplotlib inline
import numpy as np
```

```python
salario_marcelo = np.array([100,250,120,550,500])
salario_douglas = np.array([120,270,190,550,1200])
```

```python
matp.plot(salario_marcelo,ls='--',marker='s',label="Marcelo",c='purple')
matp.plot(salario_douglas,ls='--',marker='^',label="Futuro",c='red')
matp.legend()
matp.show()
```
